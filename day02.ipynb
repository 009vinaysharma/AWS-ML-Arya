#Operators

1. Arithmetic -> +, -, *, /, //, %, **
2. Relational -> >, <, >=, <=, ==, !=
3. Assignment -> +=, -=, *=, %=
4. Logical -> and, or, not
5. Bitwise -> &, |

##Arithmatic Operator
+ Addition -> +
-   Subtraction -> -
*   Multiplication -> *
+   Division -> /
+  Floor Division -> //
+   Modulus -> %
*  Exponentiation -> **

```python
# arithmetic opr

# Addition (+)
num1 = 10
num2 = 20
print(num1 + num2)  # 30

# example 1
```

```python
# Subtraction (-)
num1 = 100
num2 = 60
print(num1 - num2)  # 40

# example 1
```

```python
# Division (/)
num1 = 100
num2 = 10
print(num1 / num2)  # 10.0

# example 1
```

```python
# Modulus (%)
num1 = 100
num2 = 7
print(num1 % num2)
```

```python
# Floor Division (//)

num1 = 100
num2 = 7

print(num1 // num2)

# example 1
```

```python
# Exponentiation (**)

num1 = 5
num2 = 2

print(num1 ** num2)

# example 1
```

# Relational Operators

+ Greater Than            ->  >
+ Less Than               ->  <
+ Greater Than Equal To   ->  >=
+ Less Than Equal To      ->  <=
+ Equal To                ->  ==
+ Not Equal To            ->  !=

```python
# relational opr

# Greater Than (>)

num1 = 10
num2 = 5

print(num1 > num2)

# example 1
```

```python
# Less Than (<)

num1 = 10
num2 = 5

print(num1 < num2)

# example 1
```

```python
# Greater Than Equal To (>=)

num1 = 10
num2 = 10

print(num1 >= num2)

# example 1
```

```python
# Less Than Equal To (<=)

num1 = 10
num2 = 10

print(num1 <= num2)

# example 1
```

```python
# Equal To (==)

num1 = 10
num2 = 10

print(num1 == num2)

# example 1
```

```python
# Not Equals To (!=)

num1 = 10
num2 = 12

print(num1 != num2)

# example 1


num3 = 4
num4 = 4

print(num3 != num4)

# example 2
```

## Assignment Operators

+ Addition Assignment       -> +=
- Subtraction Assignment    -> -=
* Multiplication Assignment -> *=
+ Modulus Assignment        -> %=



Assignment Operators variable ki value ko update karne ke shortcut hote hain

```python
# Addition Assignment (+=)

num = 5
num += 5
print(num)

# example 1


num1 = 10
num2 = 20
num1 += num2
print(num1)
```

```python
# Substraction Assignment (-=)

num = 15
num -= 5
print(num)

# example 1


num1 = 20
num2 = 10
num1 -= num2
print(num1)

# example 2
```

```python
# Multiplication Assignment (*=)

num = 15
num *= 5
print(num)

# example 1


num1 = 20
num2 = 10
num1 *= num2
print(num1)

# example 2
```

```python
# Modulus Assignment (%=)

num = 15
num %= 4
print(num)

# example 1


num1 = 20
num2 = 6
num1 %= num2
print(num1)

# example 2
```
