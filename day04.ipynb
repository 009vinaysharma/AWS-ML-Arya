# Loops in Python

## For Loop
For loop ka use kisi sequence (list, tuple, string, range) ke elements ko ek-ek karke access karne ke liye kiya jata hai.

Syntax:
for variable in sequence:
    statement

Example:
for i in range(5):
    print(i)

Output:
0
1
2
3
4

## While Loop
While loop tab tak execute hota hai jab tak condition True rehti hai.

Syntax:
while condition:
    statement

Example:
i = 1
while i <= 5:
    print(i)
    i += 1

Output:
1
2
3
4
5

## Difference Between For and While Loop

| For Loop | While Loop |
|-----------|------------|
| Fixed iterations ke liye use hota hai | Condition-based iterations ke liye use hota hai |
| Sequence par iterate karta hai | Condition True hone tak chalta hai |
| Infinite loop ka chance kam hota hai | Infinite loop ka chance zyada hota hai |

# For Loop

`for` loop ek control structure hai jo sequence (jaise list, tuple, string, range, etc.) ke elements par ek-ek karke iterate karta hai. Jab sequence ke sabhi elements process ho jate hain, loop terminate ho jata hai.

### Syntax

```python
for variable in sequence:
    # statements
```

### Working

1. `for` loop sequence ke pehle element se start hota hai.
2. Har iteration mein current element variable mein store hota hai.
3. Loop ke andar ke statements execute hote hain.
4. Phir next element par move karta hai.
5. Jab sequence ke sabhi elements process ho jate hain, loop terminate ho jata hai.

### Features

- Sequence ke elements ko traverse karne ke liye use hota hai.
- Iterations ki sankhya pehle se pata ho to adhik upyogi hai.
- List, Tuple, String, Dictionary aur Range ke saath use kiya ja sakta hai.
- Code ko simple aur readable banata hai.

# While Loop

`while` loop ek control structure hai jo kisi block of code ko tab tak baar-baar execute karta hai jab tak di gayi condition `True` rahti hai. Jab condition `False` ho jati hai, loop terminate ho jata hai.

### Syntax

```python
while condition:
    # statements
```

### Working

1. Sabse pehle condition check ki jati hai.
2. Agar condition `True` hai to loop ke andar ke statements execute hote hain.
3. Statements execute hone ke baad condition dobara check ki jati hai.
4. Yeh process tab tak chalta rehta hai jab tak condition `True` rahe.
5. Jab condition `False` ho jati hai, loop terminate ho jata hai aur control loop ke baad wali statement par chala jata hai.

### Features

- Condition-based loop hai.
- Iterations ki sankhya pehle se pata na ho to use kiya jata hai.
- Infinite loop bhi bana sakta hai agar condition hamesha `True` rahe.
- Repetitive tasks ko automate karne ke liye upyogi hai.

````python
i = 1

while i <= 5:
    print(i)
    i += 1
````

````python
i = 1
sum = 0

while i <= 5:
  sum += i
  i += 1
print(sum)
````

````python
# while question
print("loop start")
while True:
  num = int(input('Enter a number: '))
  if num == 10:
    print(f"number {num} is found")
    break

print("loop end")
````

````python
# Function Calculator

def Cal(a, b, op):
    if op == "+":
        print(a + b)
    elif op == "-":
        print(a - b)
    elif op == "*":
        print(a * b)
    elif op == "/":
        print(a / b)
    else:
        print("Invalid Operator")

Cal(10, 20, '+')
Cal(10, 20, '-')
Cal(10, 20, '*')
Cal(10, 20, '/')
````
