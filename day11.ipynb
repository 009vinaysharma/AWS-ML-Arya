# Matplotlib

**Matplotlib** is a popular Python library used for data visualization. It helps create different types of charts and graphs to analyze data effectively.

### Common Types of Plots
- Line Plot
- Bar Chart
- Scatter Plot
- Histogram
- Pie Chart
- Box Plot

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Import Matplotlib library
import matplotlib.pyplot as plt

# Data
data = {
    "experience": [1, 2, 3, 4, 5],
    "salary": [100, 200, 300, 400, 500]
}

# Store data in variables
x = data["experience"]
y = data["salary"]

# Create Line Chart
plt.plot(
    x,
    y,
    marker='*',          # Marker style
    markersize=14,       # Marker size
    color='green',       # Line color
    linewidth=2,         # Line thickness
    linestyle='-'        # Line style
)

# Add Title and Labels
plt.title("Experience vs Salary")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")

# Show Grid
plt.grid(True)

# Display Graph
plt.show()
````

# Matplotlib - Line Plot

## What is a Line Plot?

A **Line Plot** is a graph that connects data points using straight lines.

It is mainly used to:
- Show trends over time.
- Compare two numerical variables.
- Analyze continuous data.

### Syntax

```python
plt.plot(x, y)
```

Where:
- `x` → Values for X-axis
- `y` → Values for Y-axis

---

## Common Parameters

| Parameter | Description |
|-----------|-------------|
| x | X-axis values |
| y | Y-axis values |
| color | Changes line color |
| marker | Adds marker on each point |
| markersize | Changes marker size |
| linestyle | Changes line style |
| linewidth | Changes line thickness |
| label | Adds legend |

---

## Marker Types

|---------|---------|

| '*' | Star |
| '+' | Plus |

---

## Line Styles

| Style | Meaning |
|--------|----------|

| ':' | Dotted Line |
| '-.' | Dash Dot |

---

## Colors

````python
import matplotlib.pyplot as plt

# Data
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

sales_2024 = [120,135,150,180,210,250,270,300,320,350,370,400]
sales_2025 = [140,155,180,210,240,290,310,340,380,420,450,500]

# Figure Size
plt.figure(figsize=(12,6))

# First Line
plt.plot(
    months,
    sales_2024,
    color="blue",
    marker="o",
    markersize=8,
    linewidth=3,
    linestyle="-",
    label="Sales 2024"
)

# Second Line
plt.plot(
    months,
    sales_2025,
    color="red",
    marker="*",
    markersize=12,
    linewidth=3,
    linestyle="--",
    label="Sales 2025"
)

# Title
plt.title(
    "Monthly Sales Comparison (2024 vs 2025)",
    fontsize=18,
    fontweight="bold",
    color="darkgreen"
)

# Axis Labels
plt.xlabel("Months", fontsize=14)
plt.ylabel("Sales (Thousands)", fontsize=14)

# Axis Limits
plt.ylim(100,550)

# Rotate X-axis Labels
plt.xticks(rotation=45)

# Grid
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5,
    alpha=0.7
)

# Background Color
plt.gca().set_facecolor("#f8f8f8")

# Data Labels
for x, y in zip(months, sales_2024):
    plt.text(x, y+5, y, ha="center", fontsize=9)

for x, y in zip(months, sales_2025):
    plt.text(x, y+5, y, ha="center", fontsize=9)

# Legend
plt.legend(
    loc="upper left",
    fontsize=11,
    shadow=True
)

# Display
plt.show()
````

# Matplotlib - Bar Chart

## What is a Bar Chart?

A **Bar Chart** is used to compare values between different categories.

Each category is represented by a rectangular bar, and the height of the bar represents its value.

### Syntax

```python
plt.bar(x, y)
```

Where:

- `x` → Categories (X-axis)
- `y` → Values (Y-axis)

---

## Common Parameters

| Parameter | Description |
|-----------|-------------|
| x | Categories |
| height | Height of bars |
| color | Bar color |
| width | Width of bars |
| edgecolor | Border color |
| linewidth | Border thickness |
| label | Legend name |

---

## Uses

- Compare student marks
- Product sales
- Company profit
- Population comparison
- Monthly expenses

````python
import matplotlib.pyplot as plt

# Data
students = ["A", "B", "C", "D", "E"]
marks = [75, 88, 65, 92, 80]

# Create Bar Chart
plt.bar(students, marks)

# Title
plt.title("Student Marks")

# Axis Labels
plt.xlabel("Students")
plt.ylabel("Marks")

# Display Graph
plt.show()
````

# Grouped Bar Chart

A **Grouped Bar Chart** is used to compare multiple categories for the same group.

In this example:
- X-axis represents Students.
- Y-axis represents Marks.
- Each color represents a different Subject.

Subjects:
- Python
- SQL
- Machine Learning
- Data Visualization

````python
import matplotlib.pyplot as plt
import numpy as np

# Students
students = ["Aman", "Riya", "Rahul", "Priya", "Karan"]

# Marks
python_marks = [88, 92, 85, 90, 95]
sql_marks = [80, 89, 84, 87, 91]
ml_marks = [90, 94, 88, 92, 96]
dv_marks = [82, 90, 86, 89, 93]

# X-axis positions
x = np.arange(len(students))

# Width of each bar
width = 0.2

# Figure Size
plt.figure(figsize=(12,6))

# Grouped Bars
plt.bar(x-0.3, python_marks, width,
        color="royalblue",
        label="Python")

plt.bar(x-0.1, sql_marks, width,
        color="green",
        label="SQL")

plt.bar(x+0.1, ml_marks, width,
        color="red",
        label="Machine Learning")

plt.bar(x+0.3, dv_marks, width,
        color="orange",
        label="Data Visualization")

# Student Names
plt.xticks(x, students)

# Labels
plt.xlabel("Students", fontsize=12)
plt.ylabel("Marks", fontsize=12)

# Title
plt.title("Student Performance in Different Subjects",
          fontsize=16,
          fontweight="bold")

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Legend
plt.legend()

# Value Labels
for i in range(len(students)):
    plt.text(x[i]-0.3, python_marks[i]+1, python_marks[i], ha='center', fontsize=8)
    plt.text(x[i]-0.1, sql_marks[i]+1, sql_marks[i], ha='center', fontsize=8)
    plt.text(x[i]+0.1, ml_marks[i]+1, ml_marks[i], ha='center', fontsize=8)
    plt.text(x[i]+0.3, dv_marks[i]+1, dv_marks[i], ha='center', fontsize=8)

# Display
plt.show()
````

# Matplotlib - Pie Chart

## What is a Pie Chart?

A **Pie Chart** is a circular graph divided into slices.

Each slice represents the proportion (percentage) of a category with respect to the whole.

### Syntax

```python
plt.pie(x)
```

Where:

- `x` → Numerical values

---

## Uses

- Market Share
- Budget Distribution
- Subject-wise Marks
- Population Percentage
- Company Revenue Share

---

## Common Parameters

| Parameter | Description |
|-----------|-------------|
| labels | Slice names |
| colors | Slice colors |
| autopct | Show percentage |
| startangle | Starting angle |
| explode | Separate slice |
| shadow | Add shadow |
| radius | Pie size |

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

plt.pie(students)

plt.title("Students Distribution")

plt.show()
````

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

plt.pie(
    students,
    labels=subjects
)

plt.title("Students Distribution")

plt.show()
````

## Show Percentage

The `autopct` parameter displays the percentage of each slice.

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

plt.pie(
    students,
    labels=subjects,
    autopct="%1.1f%%"
)

plt.title("Students Distribution")

plt.show()
````

## Change Colors

Use the `colors` parameter to assign different colors to slices.

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

colors = ["red", "blue", "green", "orange"]

plt.pie(
    students,
    labels=subjects,
    colors=colors,
    autopct="%1.1f%%"
)

plt.title("Students Distribution")

plt.show()
````

## Explode

The `explode` parameter separates one or more slices from the pie chart.

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

explode = (0.1,0,0,0)

plt.pie(
    students,
    labels=subjects,
    autopct="%1.1f%%",
    explode=explode
)

plt.title("Students Distribution")

plt.show()
````

## Shadow

Adds a shadow effect to the Pie Chart.

````python
import matplotlib.pyplot as plt

subjects = ["Python", "SQL", "ML", "Power BI"]
students = [45, 30, 15, 10]

plt.pie(
    students,
    labels=subjects,
    autopct="%1.1f%%",
    shadow=True
)

plt.title("Students Distribution")

plt.show()
````

# Matplotlib - Subplot

## What is Subplot?

A **Subplot** allows us to display multiple graphs in a single figure.

It is useful when comparing different datasets side by side.

### Syntax

```python
plt.subplot(rows, columns, index)
```

Where:

- **rows** → Number of rows
- **columns** → Number of columns
- **index** → Position of the graph

Example:

```python
plt.subplot(2,2,1)
```

Means:

- 2 Rows
- 2 Columns
- First Graph

````python
import matplotlib.pyplot as plt

# Data
months = ["Jan","Feb","Mar","Apr","May"]

sales = [120,150,180,170,200]

profit = [30,40,50,45,60]

# First Graph
plt.subplot(1,2,1)

plt.plot(months,sales,
         marker="o",
         color="blue")

plt.title("Sales")

# Second Graph
plt.subplot(1,2,2)

plt.plot(months,profit,
         marker="*",
         color="red")

plt.title("Profit")

plt.tight_layout()

plt.show()
````

````python
# sub plot
import matplotlib.pyplot as plt #

# pie chart data
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
plt.subplot(1,2,1)
plt.pie(y,labels=x)

# line chart data
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']
plt.subplot(1,2,2)
plt.plot(x,y)
plt.tight_layout()
plt.show()
````

## 4 Subplots

We can also create 4 graphs in one figure.

Layout:

```
Graph1 | Graph2
----------------
Graph3 | Graph4
```

````python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

sales = [120,150,180,170,200]
profit = [30,40,50,45,60]
expense = [70,80,85,90,95]
orders = [50,70,90,110,130]

plt.figure(figsize=(10,8))

# Graph 1
plt.subplot(2,2,1)
plt.plot(months,sales,color="blue",marker="o")
plt.title("Sales")

# Graph 2
plt.subplot(2,2,2)
plt.bar(months,profit,color="green")
plt.title("Profit")

# Graph 3
plt.subplot(2,2,3)
plt.scatter(months,expense,color="red")
plt.title("Expense")

# Graph 4
plt.subplot(2,2,4)
plt.pie(
    [40,30,20,10],
    labels=["A","B","C","D"],
    autopct="%1.1f%%"
)
plt.title("Market Share")

plt.tight_layout()

plt.show()
````

````python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

sales = [120,150,180,170,200]
profit = [30,40,50,45,60]
expense = [70,80,85,90,95]
orders = [50,70,90,110,130]

plt.figure(figsize=(10,8))

# Graph 1
plt.subplot(2,2,1)
plt.plot(months,sales,color="blue",marker="o")
plt.title("Sales")

# Graph 2
plt.subplot(2,2,2)
plt.plot(months,profit,color="green",marker="*")
plt.title("Profit")

# Graph 3
plt.subplot(2,2,3)
plt.plot(months,expense,color="red",marker=".")
plt.title("Expense")

# Graph 4
plt.subplot(2,2,4)
plt.pie(
    [40,30,20,10],
    labels=["A","B","C","D"],
    autopct="%1.1f%%"
)
plt.title("Market Share")

plt.tight_layout()

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Data


# Line Plot
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [120,150,180,170,210,250]

# Bar Chart
profit = [30,45,50,42,60,75]

# Pie Chart
products = ["Laptop","Mobile","Tablet","Accessories"]
market_share = [40,30,20,10]

# Multiple Bar Chart
students = ["A","B","C","D","E"]

python_marks = [85,90,88,92,95]
sql_marks = [80,86,84,89,91]

x = np.arange(len(students))
width = 0.35


# Figure

plt.figure(figsize=(14,10))


# 1. Line Plot

plt.subplot(2,2,1)

plt.plot(
    months,
    sales,
    color="blue",
    marker="o",
    linewidth=3,
    label="Sales"
)

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()


# 2. Bar Chart

plt.subplot(2,2,2)

plt.bar(
    months,
    profit,
    color="green",
    edgecolor="black"
)

plt.title("Monthly Profit")
plt.xlabel("Months")
plt.ylabel("Profit")
plt.grid(axis="y")


# 3. Pie Chart
plt.subplot(2,2,3)

plt.pie(
    market_share,
    labels=products,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True
)

plt.title("Market Share")

# 4. Multiple Bar Chart

plt.subplot(2,2,4)

plt.bar(
    x-width/2,
    python_marks,
    width,
    color="royalblue",
    label="Python"
)

plt.bar(
    x+width/2,
    sql_marks,
    width,
    color="orange",
    label="SQL"
)

plt.xticks(x, students)

plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.legend()


plt.tight_layout()

# Display
plt.show()
````
