# Fancy Indexing

Fancy Indexing NumPy ka ek advanced indexing method hai jisme hum **list ya array of indices** ka use karke multiple elements ko ek saath access karte hain.

### Syntax

```python
array[[index1, index2, index3]]
```

### 2D Syntax

```python
array[[row1, row2], [col1, col2]]
```

Yeh pair-wise indexing karta hai.

```
array[[0,1],[2,1]]

↓

array[0,2]
array[1,1]
```

### Uses

- Multiple elements ko ek saath access karna
- Specific rows select karna
- Specific columns select karna
- Random indexing

````python
# fancy indexing
import numpy as np
arr = np.array([10,20,30,40,50,60])
print(arr)
arr_f = arr[[0,2]] # 0 index value, 2 index value
print(arr_f)
````

````python
import numpy as np

arr = np.array([10,20,30,40,50,60])

print(arr[[0,2,5]])
````

````python
import numpy as np

arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

arr_f = arr[[0,2],[1,2]]
arr_f1 = arr[[0,1],[2,2]]
print(arr_f)
print(arr_f1)
````

````python
import numpy as np
# selecting multiple rows
arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

print(arr[[0,2]])

# selecting multiple columns
import numpy as np

arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

print(arr[:, [0,2]])
````

# np.where() Condition

`np.where()` NumPy ka ek function hai jo condition ke basis par values ya indices return karta hai.

## Syntax

```python
np.where(condition)
```

Returns the index positions where the condition is **True**.

### Syntax (Ternary Operator)

```python
np.where(condition, value_if_true, value_if_false)
```

- **condition** → Check ki jane wali condition
- **value_if_true** → Agar condition True ho
- **value_if_false** → Agar condition False ho

## Uses

- Find indices
- Replace values
- Apply conditions on arrays

````python
# where
import numpy as np
arr = np.array([10,20,3,4,90,100])
print(arr)
arr_w = np.where(arr > 9,'True','False')
print(arr_w)
````

````python
arr = np.array([5,6,7,8,11,24,16,19,25])
print(arr)
arr_w = np.where((arr>10) & (arr<20),arr*2,arr+2)
print(arr_w)

````

````python
import numpy as np

arr = np.array([10,20,30,40,50])

print(np.where(arr > 25))
````

````python
import numpy as np

arr = np.array([10,20,30,40,50])

print(np.where(arr > 25, arr+100, arr))
````

````python
import numpy as np

arr = np.array([10,20,30,40,50])

print(np.where(arr > 25, "True", "False"))
````

````python
import numpy as np

arr = np.array([1,2,3,4,5,6])

print(np.where(arr % 2 == 0, "Even", "Odd"))
````

````python
import numpy as np

arr = np.array([-10,20,-30,40,-50])

print(np.where(arr < 0, -arr, arr))
````

# Adding and Removing Data in NumPy

NumPy provides functions to add (join) and remove elements, rows, and columns from arrays.

## 1. Concatenate

Used to join two arrays.

```python
np.concatenate((arr1, arr2))
```

---

## 2. Add Row

Used to add a new row to a 2D array.

```python
np.vstack((original, new_row))
```

---

## 3. Add Column

Used to add a new column to a 2D array.

```python
np.hstack((original, new_col))
```

---

## 4. Delete Element

Used to remove an element from a 1D array or a row/column from a 2D array.

```python
np.delete(array, index)
```

````python
# 3D Concatenate
import numpy as np

arr1 = np.array([
    [[1,2],
     [3,4]]
])

arr2 = np.array([
    [[5,6],
     [7,8]]
])

# axis = 0 (Depth/Block Wise)
result0 = np.concatenate((arr1, arr2), axis=0)
print("Axis = 0")
print(result0)

# axis = 1 (Row Wise)
result1 = np.concatenate((arr1, arr2), axis=1)
print("\nAxis = 1")
print(result1)

# axis = 2 (Column Wise)
result2 = np.concatenate((arr1, arr2), axis=2)
print("\nAxis = 2")
print(result2)
````

# Statistical Functions in NumPy

Statistical functions are used to perform mathematical and statistical calculations on NumPy arrays.

## Functions

### 1. np.sum()
Returns the sum of all elements.

### 2. np.mean()
Returns the average of all elements.

**Formula:**
Mean = Sum of Elements / Number of Elements

### 3. np.median()
Returns the middle value after sorting the data.

### 4. np.mode()
Returns the most frequently occurring value.
(NumPy has no direct `mode()` function. It is available in `scipy.stats`.)

### 5. np.min()
Returns the smallest element.

### 6. np.max()
Returns the largest element.

### 7. np.var()
Returns the variance of the data.

### 8. np.std()
Returns the standard deviation.

### 9. np.prod()
Returns the product of all elements.

### 10. np.cumsum()
Returns the cumulative (running) sum.

### 11. np.cumprod()
Returns the cumulative (running) product.

### 12. np.argmin()
Returns the index of the minimum value.

### 13. np.argmax()
Returns the index of the maximum value.

### 14. np.abs()
Returns the absolute value.

### 15. np.unique()
Returns only unique elements.

### 16. np.percentile()
Returns the value at a given percentile.

### 17. np.quantile()
Returns the value at a given quantile.

### 18. np.ptp()
Returns the range (Maximum − Minimum).

### 19. np.any()
Returns True if at least one element is True.

# Handling Missing Values

Missing values are values that are not available in a dataset.
In NumPy, missing values are represented using **np.nan (Not a Number)**.

## Methods

- `np.isnan()` → Check missing values
- `np.nan_to_num()` → Replace missing values
- `np.nanmean()` → Mean ignoring missing values
- `np.nansum()` → Sum ignoring missing values
- `np.nanmax()` → Maximum ignoring missing values
- `np.nanmin()` → Minimum ignoring missing values

````python
import numpy as np

arr = np.array([10,20,np.nan,40,50])

print(arr)

print(np.isnan(arr))

print(np.nan_to_num(arr, nan=0))
````

````python
import numpy as np
#mean ignoring mising values
arr = np.array([10,20,np.nan,40,50])

print(np.nanmean(arr))
````

````python
import numpy as np
# sum ignoring missing values
arr = np.array([10,20,np.nan,40,50])

print(np.nansum(arr))
````

````python
import numpy as np

arr = np.array([10,20,np.nan,40,50])
# Maximum & Minimum Ignoring Missing Values
print("Maximum :", np.nanmax(arr))
print("Minimum :", np.nanmin(arr))
````

````python
import numpy as np

# Creating array with missing values
arr = np.array([10,20,np.nan,40,50,np.nan,70])

print("Original Array :")
print(arr)

# Check missing values
print("\nMissing Values (True = Missing)")
print(np.isnan(arr))

# Count missing values
print("\nTotal Missing Values :")
print(np.sum(np.isnan(arr)))

# Replace missing values with 0
print("\nReplace Missing Values with 0")
print(np.nan_to_num(arr, nan=0))

# Replace missing values with 100
print("\nReplace Missing Values with 100")
print(np.nan_to_num(arr, nan=100))

# Mean ignoring missing values
print("\nMean :", np.nanmean(arr))

# Sum ignoring missing values
print("Sum :", np.nansum(arr))

# Maximum ignoring missing values
print("Maximum :", np.nanmax(arr))

# Minimum ignoring missing values
print("Minimum :", np.nanmin(arr))
````
