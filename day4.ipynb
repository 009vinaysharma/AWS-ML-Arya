#Methods Of Lists
Method	Use
+ append() ->	End me item add karta hai
+ insert() ->	Kisi position par item add karta hai
+ extend() ->	Dusri list ke items add karta hai
+ remove() ->	Value ko delete karta hai
+ pop() ->	Index se item delete karke return karta hai
+ clear() ->	Puri list empty kar deta hai
+ index() ->	Item ka index batata hai
+ count() ->	Item kitni baar hai batata hai
+ sort() ->	List ko sort karta hai
+ reverse() ->	List ko ulta karta hai
+ copy() ->	List ki copy banata hai

```python
lst = [10, 20, 30]

lst.append(40)
print(lst)       # [10, 20, 30, 40]

lst.insert(1, 15)
print(lst)       # [10, 15, 20, 30, 40]

lst.remove(20)
print(lst)       # [10, 15, 30, 40]

lst.pop()
print(lst)       # [10, 15, 30]

lst.sort()
print(lst)

lst.reverse()
print(lst)

print(lst.count(10))
print(lst.index(15))
```

# Loops in Python

## For Loop
For loop ka use kisi sequence (list, tuple, string, range) ke elements ko ek-ek karke access karne ke liye kiya jata hai.

Syntax:
for variable in sequence:
    statement

Example:
for i in range(5):
    print(i)

Output:
0
1
2
3
4

## While Loop
While loop tab tak execute hota hai jab tak condition True rehti hai.

Syntax:
while condition:
    statement

Example:
i = 1
while i <= 5:
    print(i)
    i += 1

Output:
1
2
3
4
5

## Difference Between For and While Loop

| For Loop | While Loop |
|-----------|------------|
| Fixed iterations ke liye use hota hai | Condition-based iterations ke liye use hota hai |
| Sequence par iterate karta hai | Condition True hone tak chalta hai |
| Infinite loop ka chance kam hota hai | Infinite loop ka chance zyada hota hai |

# Function in Python

Function code ka ek block hota hai jo kisi specific task ko perform karta hai. Function ko baar-baar call kiya ja sakta hai.

## Syntax

def function_name():
    statement



## Types of Functions in Python

- ### 1. Built-in Functions
    Python ke predefined functions.

   Examples:
    + print()
    + len()
    + input()
    + type()

- ### 2. User-defined Functions
    User khud function banata hai.

    Example:

    def greet():
    
    print("Hello")

    greet()




### -> Function Arguments

def add(a, b):
    print(a + b)

add(10, 20)

Output:
30

### -> Return Statement

def square(n):
    return n * n

print(square(5))

Output:
25

## Advantages of Functions

1. Code Reusability
2. Easy Debugging
3. Code Readability
4. Time Saving
