# Exception Handling

**Exception Handling** ek technique hai jo program ke execution ke dauran aane wali errors (exceptions) ko handle karne ke liye use ki jati hai. Isse program crash hone ke bajay smoothly execute hota rehta hai.

### Syntax

```python
try:
    # code that may cause an exception
except:
    # code to handle the exception
```

### Working

1. `try` block mein aisa code likha jata hai jahan error aane ki sambhavana ho.
2. Agar error aata hai, to control `except` block mein chala jata hai.
3. `except` block error ko handle karta hai.
4. Agar error nahi aata, to `except` block execute nahi hota.

### Keywords Used

- `try` → Error-prone code likhne ke liye
- `except` → Error handle karne ke liye
- `else` → Error na aaye to execute hota hai
- `finally` → Hamesha execute hota hai

### Features

- Program ko crash hone se bachata hai.
- Runtime errors ko handle karta hai.
- Program ki reliability badhata hai.
- User-friendly error messages provide karta hai.

````python
try:
  print("try part")
except:
  print("except part")
````

````python
try:
  a = 10
  b = 0
  print(a/b)
except:
  print("zero division error")
````

````python
a = 10
b = 0

try:
  print(a/b)
except:
  print("zero division error")
````

````python
try:
  print("hello" * 10)
except TypeError:
  print('type error')
except ZeroDivisionError:
  print('zero division error')
````

````python
try:
  print("try part")
except:
  print("except part")
finally:
  print("finally part")
````

````python
try:
  print("try part")
  print(10/0)
except:
  print("except part")
finally:
  print("finally part")
````

````python

try:
  a = 10
  b = 0
  print(a/b)
  print("try part")

except:
  print("except part")
finally:
  print("finally part")
````

````python
try:
    num = int(input("Enter a number: "))
    print(num)
except ValueError:
    print("Please enter a valid number")
````

````python
try:
    a = 10
    b = 2
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
finally:
    print("Program ended")
````

# Lambda Function

**Lambda Function** Python mein ek anonymous (naam-rahit) function hota hai jo ek hi expression ko evaluate karta hai aur uska result return karta hai. Ise chhote aur simple functions ko likhne ke liye use kiya jata hai.

### Syntax

```python
lambda arguments : expression
```

### Working

1. `lambda` keyword se function create kiya jata hai.
2. Arguments function ke input hote hain.
3. Expression evaluate hota hai.
4. Expression ka result automatically return ho jata hai.
5. Is function ka koi naam nahi hota, isliye ise Anonymous Function bhi kehte hain.

### Features

- Anonymous (naam-rahit) function hota hai.
- Sirf ek expression likh sakte hain.
- Automatic return karta hai.
- Chhote aur simple functions ke liye upyogi hai.
- `map()`, `filter()`, aur `reduce()` ke saath adhik use hota hai.

````python
add = lambda a, b: a + b

print(add(10, 20))
````

````python
square = lambda x: x * x

print(square(5))
````

````python
maximum = lambda a, b: a if a > b else b

print(maximum(10, 20))
````

````python
add = lambda x, y: x + y
print(add(10, 20))

cal = lambda *a: eval(f"{a[0]} {a[-1]} {a[1]}")
print(cal(10, 20, '*'))
````

````python
num = lambda n: [i for i in range(1, n+1)]

print(num(5))
````

````python
even = lambda x: x % 2 == 0

print(even(10))
````

# Membership Operators

**Membership Operators** ka use yeh check karne ke liye kiya jata hai ki koi value kisi sequence (String, List, Tuple, Set, Dictionary) mein present hai ya nahi.

### Types of Membership Operators

1. `in`
2. `not in`

### Syntax

```python
value in sequence

value not in sequence
```

### Working

1. `in` operator check karta hai ki value sequence mein present hai ya nahi.
2. Agar value mil jati hai to `True` return karta hai, warna `False`.
3. `not in` operator `in` ka opposite hai.
4. Agar value present nahi hai to `True` return karta hai.

### Features

- String, List, Tuple, Set aur Dictionary ke saath use hota hai.
- Result hamesha Boolean (`True` ya `False`) hota hai.
- Searching aur validation ke liye upyogi hai.

````python
nums = [10, 20, 30, 40]

print(20 in nums)
print(50 not in nums)
````

````python
name = "Vinay"

print("V" in name)
print("z" not in name)
````

````python
d = {"name": "Vinay", "age": 20}

print("name" in d)
print("city" not in d)
````

# List Comprehension

**List Comprehension** Python mein ek concise (short) aur efficient tarika hai jisse existing iterable (List, Tuple, String, Range, etc.) se nayi list banayi jati hai.

### Syntax

```python
[expression for item in iterable]
```

### Syntax with Condition

```python
[expression for item in iterable if condition]
```

### Working

1. Iterable ke har element par loop chalta hai.
2. Har element par expression apply hota hai.
3. Agar condition di gayi ho to pehle condition check hoti hai.
4. Satisfy hone wale elements ko nayi list mein add kiya jata hai.
5. Result ek nayi list ke roop mein return hota hai.

### Features

- Code ko short aur readable banata hai.
- Traditional `for` loop ka concise alternative hai.
- Filtering aur transformation dono kar sakta hai.
- Faster aur efficient hota hai.

````python
# list comprehension
[i for i in range(5)]
````

````python
# 2 table list comprehension
[i * 2 for i in range(1,11)]
````

````python
# 3 table list comprehension
[i * 3 for i in range(1,11,)]
````

````python
# 4 table list comprehension
[i * 4 for i in range(1,11,)]
````

````python
squares = [i**2 for i in range(1, 6)]
print(squares)
````

````python
#Square of Numbers
[i**2 for i in range(int(input("Enter a Number")))]
````

````python
[i/2 for i in range(8)]
````

````python
# Even Numbers
[i for i in range(1,11) if i % 2 ==0]

````

````python
# odd numbers
[i for i in range(1,11) if (i % 2)-1 ==0]
````

# Module

**Module** Python file hoti hai jisme functions, classes aur variables ka collection hota hai. Module ka use code ko reusable aur organized banane ke liye kiya jata hai.

### Syntax

```python
import module_name
```

### Working

1. Module ek `.py` file hoti hai.
2. Isme functions, variables aur classes define kiye jate hain.
3. `import` keyword ki madad se module ko program mein use kiya jata hai.
4. Module ke functions aur variables ko dot (`.`) operator se access kiya jata hai.

### Features

- Code reusability provide karta hai.
- Program ko modular aur organized banata hai.
- Code duplication kam karta hai.
- Maintenance aur debugging ko easy banata hai.

````python
import math
````

````python
print(m.pi)
````

````python
# module import

# Example 1
import math
print(math.pi)
print(math.factorial(5))

# Example 2
import math as m
print(m.pi)
print(m.factorial(5))

# Example 3
from math import pi,factorial
print(pi)
print(factorial(5))

# Example 4
from math import *
print(pi)
print(factorial(5))
print(pow(5,3))
````
