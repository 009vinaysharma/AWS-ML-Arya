# Encapsulation in Python

## Definition
Encapsulation is an OOP concept that binds data (variables) and methods (functions) into a single unit (class) and restricts direct access to data.

It helps in:
- Data Hiding
- Security
- Better Control over Data

## Example

```python
class Student:
    def __init__(self, name, marks):
        self.name = name          # Public Variable
        self.__marks = marks      # Private Variable

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.__marks)

s = Student("Vinay", 90)
s.display()
```

## Output

```
Name: Vinay
Marks: 90
```

## Private Variable

```python
self.__marks
```

Variables starting with `__` (double underscore) are private and cannot be accessed directly outside the class.

```python
print(s.__marks)
```

Output:

```
AttributeError
```

## Advantages

- Protects data from unauthorized access.
- Improves security.
- Makes code more maintainable.
- Provides data hiding.

# Method Overloading vs Method Overriding

| Method Overloading | Method Overriding |
|-------------------|------------------|
| Same method name with different parameters. | Same method name and same parameters in parent and child class. |
| Achieved within the same class. | Achieved using inheritance. |
| Python does not support true method overloading directly. | Python fully supports method overriding. |
| Increases readability. | Provides specific implementation in child class. |

## Method Overloading Example

```python
class Demo:
    def add(self, a, b, c=0):
        print(a + b + c)

d = Demo()

d.add(10, 20)
d.add(10, 20, 30)
```

### Output

```
30
60
```

## Method Overriding Example

```python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        print("Dog Barks")

d = Dog()
d.sound()
```

### Output

```
Dog Barks
```

## Definition (English)

### Method Overloading

### Method Overriding
Method overriding is the feature in which a child class provides a specific implementation of a method already defined in the parent class.

````python
class Student:
    def info(self, name, age, marks="Absent"):
        if age is None:
            print("Name:", name,"Age:",age)
        else:
            print("Name:", name, "Age:", age, "Marks", marks)

s = Student()

s.info("Vinay",20)
s.info("Vinay",20,90)
s.info("raj",23,9)
````

````python
# To achieve similar functionality to overloading in Python, use default arguments.
def Student123(name, roll_no, age=None):
  if age is not None:
    print(name, roll_no, age)
  else:
    print(name, roll_no)

Student123("rajendra", 123, 20) # 3 arguments
Student123("rajendra", 123) # 2 arguments
````
