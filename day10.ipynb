# Pandas

## Aim
To study the Pandas library and understand its features for data manipulation and analysis.

## Introduction
Pandas is an open-source Python library used for data analysis and data manipulation. It provides fast, flexible, and efficient data structures for working with structured data.

## Features of Pandas
- Easy handling of missing data.
- Fast and efficient data analysis.
- Supports reading and writing data in various formats such as CSV, Excel, JSON, and SQL.
- Powerful data filtering, sorting, and grouping.
- Built-in statistical and data analysis functions.
- Data cleaning and preprocessing support.
- Time series analysis support.

## Applications
- Data Analysis
- Data Cleaning
- Data Visualization
- Machine Learning
- Business Intelligence
- Financial Data Analysis

## Advantages
- Easy to learn and use.
- High performance.
- Integrates well with NumPy and Matplotlib.
- Supports large datasets.
- Provides powerful data manipulation tools.

## Conclusion
Pandas is one of the most important Python libraries for data analysis. It simplifies working with structured data and is widely used in data science, machine learning, and analytics.

````python
import pandas as pd
l = [10,20,30,40,50] # list
df = pd.Series(l) # use list in pd
print(df)


# example data
df = pd.Series(data=l)
print(df)

# example index
df = pd.Series(data=l , index=['nikhil','devendra','raj','hello','unknown'])
print(df)

# example key = index | value = value
d = {'nikhil':10,'devendra':20,'raj':30,'hello':40,'unknown':50}
df = pd.Series(data=d)
print(df)
````

````python
l = [10,20,30,40,50,60] #List
df = pd.Series(l)
df.iloc[0] = 99
print(df)
print(l)
````

````python
import pandas as pd
data = {
    "name":['nikhli','raj','devendra','hello'],
    "age":[10,20,30,40]
}
# 1D representation
df = pd.Series(data)
print(df)

# 2D representation
df = pd.DataFrame(data)
print(df)

````

````python
import pandas as pd
data =[['devendra',10],['nikhil',20],['raj',30],['hello',40]]
df = pd.DataFrame(data)
print(df)

# Add Column
df = pd.DataFrame(data,columns=['name','age'])
print(df)

# Add indexing name
df = pd.DataFrame(data,index=['a','b','c','d'])
print(df)
````
