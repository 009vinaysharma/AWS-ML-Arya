# Object-Oriented Programming (OOP)

## What is OOP?

Object-Oriented Programming (OOP) is a programming paradigm that organizes software around **objects** rather than functions.

An object contains:

- **Attributes (Data)** → Properties of the object
- **Methods (Behavior)** → Actions the object can perform

### Example

A Car has:

- Brand
- Color
- Speed

And can:

- Start
- Stop
- Accelerate

# Four Pillars of OOP

## 1. Encapsulation

### Definition

Encapsulation means wrapping data and methods together and restricting direct access to data.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

### Benefits

- Data Security
- Controlled Access
- Better Maintenance

---

## 2. Inheritance

### Definition

Inheritance allows one class to acquire properties and methods from another class.

```python
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    pass
```

### Benefits

- Code Reusability
- Reduced Redundancy

### Types of Inheritance

1. Single Inheritance
2. Multilevel Inheritance
3. Multiple Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

---

## 3. Polymorphism

### Definition

Polymorphism means **one interface, many forms**.

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

### Output

```
Bark
Meow
```

### Method Overriding

```python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        print("Bark")
```

---

## 4. Abstraction

### Definition

Abstraction hides implementation details and shows only essential features.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")
```

### Benefits

- Hides Complexity
- Improves Security
- Easy Maintenance

# Inheritance in OOP

Inheritance is a mechanism where one class acquires the properties and methods of another class.

## Advantages
- Code reusability
- Easy maintenance
- Extensibility

### Example
```python
class Animal:
    pass

class Dog(Animal):
    pass




````python
class Animal:
    pass

class Dog(Animal):
    pass
````

````python
class Section:
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no

obj = Section("nikhil",123)
print(obj.name)
print(obj.roll_no)

class Section_A(Section):
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no

obj = Section_A("nikhil",123)
print(obj.name)
print(obj.roll_no)


class Section_B(Section):
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no

obj = Section_B("rajendra",321)
print(obj.name)
print(obj.roll_no)
````

````python
class Section:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

class Section_A(Section):
    def __init__(self, name, roll_no):
        super().__init__(name, roll_no)

class Section_B(Section):
    def __init__(self, name, roll_no):
        super().__init__(name, roll_no)

obj1 = Section_A("nikhil", 123)
obj2 = Section_B("rajendra", 321)

print(obj1.name, obj1.roll_no)
print(obj2.name, obj2.roll_no)
````

````python
class Animal:
    def __init__(self, Bark):
        self.bark = Bark

class Dog(Animal):
    def __init__(self, Bark):
        super().__init__(Bark)

obj = Dog("Woof Woof")
print(obj.bark)
````

````python
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def getData(self):
        print(self.name, self.roll_no)


class Address(Student):   # Inheriting Student
    def __init__(self, name, roll_no, city, state):
        super().__init__(name, roll_no)  # Call parent constructor
        self.city = city
        self.state = state

    def getAddress(self):
        print(self.city, self.state)


# Object creation
obj = Address("John", 101, "New York", "NY")

obj.getData()      # Inherited method
obj.getAddress()   # Child class method
````

# super() in Python

- `super()` is a built-in function used to access the parent class methods and constructor from a child class.
- It helps in code reusability and avoids rewriting parent class code.
- Commonly used to call the parent class constructor using `super().__init__()`.

## Example

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

d = Dog("Tommy")
print(d.name)


````python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

d = Dog("Tommy")
print(d.name)
````

````python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        print("Dog Barks")

d = Dog()
d.sound()                                #Without super(), parent class ka method call nahi hoga.
````

````python
# method call example
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog Barks")

d = Dog()
d.sound()
````

````python
class Vinay:
    def __init__(self, bike):
        self.bike = bike

    def usebike(self):
        print("bike used by: ", self.bike)


class Rajendra(Vinay):
    def __init__(self, name, bike):
        self.name = name

    def usename(self):
        print("name is: ", self.name)


u = Rajendra("rajendra janigd", "tvs by rajendra")
u.usename()
````

````python
class Vinay:
   def __init__(self,bike):
       self.bike = bike # Initialize self.bike

   def usebike(self): # Moved to class level
     print("bike used by: ",self.bike)

class Rajendra(Vinay):
  def __init__(self,name,bike):
    super().__init__(bike) # Calls Vinay's constructor to set self.bike
    self.name = name       # Correct: Assign 'name' parameter to 'self.name'
    # Removed: self.bike = name, as it was overwriting the bike attribute with the name

  def usename(self):
    print("name is: ",self.name)

u = Rajendra("rajendra janigd","tvs by rajendra")
u.usename()
print(u.bike) # Added print for u.bike to show its value
````
